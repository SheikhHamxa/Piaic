```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"datasets_19_420_Iris.csv")
```

```python
df
```

```python
df.info()
```

```python
df["Species"].value_counts()
```

```python
dic = {
    'Iris-versicolor': 1,
    'Iris-setosa': 2,
    'Iris-virginica': 3,
}
df['Specie-Value'] = df.Species.map(dic)
```

```python
df.tail()
```

```python
df
```

```python
df = df.sample(frac=1)  #shuffle data for good accuracy
df.head()
```

```python
x=df.loc[:,'SepalLengthCm':'PetalWidthCm']
x.head()
```

```python
y=df['Specie-Value']
y.head()
```

```python
y = pd.get_dummies(y)
y.head()
```

```python
y.drop_duplicates()
```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, SGD
```

```python
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(4,)))
model.add(Dense(64, activation='relu', input_shape=(4,)))
model.add(Dense(3, activation='softmax'))
```

```python
model.summary()
```

```python
# select loss function based on your problems

model.compile(optimizer=SGD(lr=0.01),
             loss='categorical_crossentropy',
             metrics=['acc'])
```

```python
model.fit(x, y, epochs=30)
```

```python
df['predict'] = y.columns.take(np.argmax(model.predict(df.loc[:,'SepalLengthCm':'PetalWidthCm']), axis=1))
```
